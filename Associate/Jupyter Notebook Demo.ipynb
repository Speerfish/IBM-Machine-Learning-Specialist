## Jupyter Notebook Keyboard Shortcuts ##

Markdown formatting guide : https://www.markdownguide.org/basic-syntax/

- **2 spaces and a return** create a line break.
- **Shift + Enter** executes the current cell and moves to the next one.  
- **Ctrl + Enter** executes the current cell but does not move to the next one.  
- **Ctrl + Shift + -** will split the current cell into two from where your cursor is.  
- **Shift + J or Shift + Down** selects the next cell in a downwards direction.  
- **Shift + K or Shift + Up** select cells in an upwards direction.  
- **Shift + M** merges multiple cells.  

**Esc** takes you in to command mode. Clicking in a cell or pressing ** Enter ** takes you out.
While in command mode:
- **A** inserts a new cell above.
- **B** inserts a new cell below.
- **D D** deletes current cell.
- **Y** switch cell to code mode.
- **M** switch cell to Markdown format.

## Let's use the Matplotlib library to make a plot ##

```python
# Creating a range from 0 to 2 with a step of 0.0001
x_axis = np.arange(0.0, 2.0, 0.0001)
y_axis = 2 + np.sin(2 * np.pi * x_axis)

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis,color="green")
ax.plot(x_axis, 2*y_axis,color="red")

ax.set(xlabel='Time (s)', ylabel='Voltage (mV)',title='Simple Plot')
ax.grid()

plt.show()
```
